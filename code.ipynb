**Model Selection**

1.Define features (X) and the target (Y).

2.Apply feature scaling.

3.Split the data into training and test sets.

4.Choose a machine learning model.

5.Fine-tune the model's hyperparameters.







```python
# Assign features (X) and target (y)
X = data.drop(columns=["fetal_health"])
y = data["fetal_health"]

# Standardize the features and display summary statistics
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X.describe().T
```

```python
# Splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Random Forest**

```python
# Define parameters to test for the Random Forest model
param_grid = {
'n_estimators': [100, 150, 200, 500, 700, 900],
'max_depth': [4, 6, 8, 12, 14, 16],
}

# Use GridSearchCV to find the best combination of parameters
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

#Best parameter values
best_params = grid_search.best_params_

#Random Forest model with the best parameters and train it
best_RF_model = RandomForestClassifier(**best_params)
best_RF_model.fit(X_train, y_train)

# Testing
from sklearn.metrics import accuracy_score

predictions = best_RF_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, mean_squared_error

# Calculate performance metrics
metrics = {
    "Accuracy": accuracy_score(y_test, predictions),
    "Recall": recall_score(y_test, predictions, average="weighted"),
    "Precision": precision_score(y_test, predictions, average="weighted"),
    "F1 Score": f1_score(y_test, predictions, average="micro"),
    "Mean Squared Error": mean_squared_error(y_test, predictions)
}

# Display Random Forest results
print("********* Random Forest Results *********")
for metric, value in metrics.items():
    print(f"{metric:<20}: {value:.4f}")
```

**Logistic Regression**

```python
import warnings
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

# Logistic Regression model
logreg = LogisticRegression()

# Hyperparameter grid for tuning
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Regularization strength
    'penalty': ['l1', 'l2'],              # Type of penalty
    'solver': ['liblinear', 'saga']       # Solvers that support l1 penalty
}

# GridSearchCV object
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy')

warnings.filterwarnings("ignore", category=UserWarning)

# Fit the model to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters found by GridSearchCV
print("Best parameters: ", grid_search.best_params_)
```

```python
import warnings
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, mean_squared_error

# Predictions using the best hyperparameters
best_model = grid_search.best_estimator_
predictions = best_model.predict(X_test)

# Calculate performance metrics
metrics = {
    "Accuracy": accuracy_score(y_test, predictions),
    "Recall": recall_score(y_test, predictions, average="weighted"),
    "Precision": precision_score(y_test, predictions, average="weighted"),
    "F1 Score": f1_score(y_test, predictions, average="micro"),
    "Mean Squared Error": mean_squared_error(y_test, predictions)
}

# Display Logistic Regression results
print("********* Logistic Regression Results *********")
for metric, value in metrics.items():
    print(f"{metric:<20}: {value:.4f}")
```
